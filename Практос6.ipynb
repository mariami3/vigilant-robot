Диаграмма рассеяния используется для оценки взаимосвязи двух переменных,
значения которых откладываются по разным осям. Для её построения
используется функция plt.scatter(), аргументами которой выступают переменные с
дискретными значениями в данной диаграмме выступали значения счета по математике.

```python
import numpy as np 
import matplotlib.pyplot as plt 

np.random.seed(45)
x = np.random.rand(100)
y_positive = 2 * x + 1 + 0.1 * np.random.randn(100)
y_negative = -2 * x + 1 + 0.1 * np.random.randn(100)
y_no_relation = np.random.randn(100)

plt.figure(figsize=(20, 5))

plt.subplot(131)
plt.scatter(x, y_positive, color='purple', alpha=0.6)
plt.title('Положительная связь')

plt.subplot(132)
plt.scatter(x, y_negative, color='pink', alpha=0.6)
plt.title('Отрицательная связь')

plt.subplot(133)
plt.scatter(x, y_no_relation, color='blue', alpha=0.6)
plt.title('Отсутствие связь')

plt.show()
```

Здесь используется также диаграмма рассеяния но с тремя связями такими как положительными, отрицательными и отсутствием связи за значение также были взяты данные о счете по математике

```python
data = np.random.randn(100)

plt.hist(data, bins=10, color='pink', edgecolor='blue')
plt.title('Пол и раса')
plt.xlabel('Раса')
plt.ylabel('Группа')
plt.show()
```

Гистограмма представляет собой визуальный способ отображения распределения
данных. Она позволяет оценить, как часто появляются определенные значения в
наборе данных и как эти значения распределены. Гистограммы особенно полезны
для исследования формы распределения, выявления выбросов, определения
центральной тенденции и других характеристик данных. Тут также были взяты значения о гендере и группе.
